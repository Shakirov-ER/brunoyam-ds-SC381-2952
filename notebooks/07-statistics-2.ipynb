## Основные стат тесты

### T-test для сравнения двух репрезентативных количественных выборок

```python
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
```

T-тест (t-test) является одним из наиболее распространенных статистических тестов, который используется для сравнения средних значений двух независимых выборок. Он основан на распределении Стьюдента (Student's t-distribution), которое позволяет оценивать средние значения выборок, когда известны только их выборочные средние, стандартные отклонения и размеры выборок.

```python
# генерируем данные для выборки A
sample_a = np.random.normal(loc=10, scale=2, size=30)

# генерируем данные для выборки B
sample_b = np.random.normal(loc=12, scale=2, size=30)

# выполняем T-тест для двух независимых выборок
t_statistic, p_value = stats.ttest_ind(sample_a, sample_b)

# выводим результаты
print("T-статистика:", t_statistic)
print("P-значение:", p_value)

# проверяем значимость различий
alpha = 0.05
if p_value < alpha:
    print("Различия между выборками статистически значимы")
else:
    print("Нет статистически значимых различий между выборками")
```

#### T-тест (t-test) имеет следующие ограничения:

Нормальность распределения средних: T-тест предполагает, что средневыборочные ваших данных имеют нормальное распределение. Иными словами: выборки должны быть репрезентативными

Ограниченный размер выборки: T-тест работает лучше для больших выборок. При маленьких размерах выборок, результаты могут быть менее точными и ненадежными.

Дисперсии выборок должны быть равны: T-тест предполагает, что дисперсии двух выборок одинаковы. Если дисперсии значительно отличаются друг от друга, то результаты теста могут быть неточными.

Необходимость независимости выборок: T-тест может использоваться только для сравнения двух независимых выборок. Если выборки зависимы, то нужно использовать другие тесты, например, парный t-тест.

Выбросы: T-тест чувствителен к выбросам. Если в выборке присутствуют выбросы, то результаты теста могут быть недостоверными.

Множественные сравнения: T-тест предназначен для сравнения только двух выборок. При сравнении более двух выборок необходимо использовать другие методы, такие как ANOVA и post-hoc тесты.

Важно учитывать эти ограничения при использовании T-теста и оценивать надежность результатов теста в контексте специфики данных и задачи исследования.

#### Мощность теста (statistical power) - 
это вероятность того, что статистический тест найдет статистически значимые различия между двумя группами, если эти различия действительно существуют. Мощность теста зависит от нескольких факторов, включая уровень значимости (α), размер выборки (n), величину различия между группами (эффект размера), дисперсию данных, а также тип статистического теста, используемого для анализа данных.

Мощность теста может быть выражена как 1-β, где β обозначает вероятность ошибки второго рода, то есть вероятность принять нулевую гипотезу, когда она на самом деле неверна. Чем выше мощность теста, тем меньше вероятность ошибки второго рода, и тем более точными будут результаты теста.

Чтобы увеличить мощность теста, можно использовать следующие стратегии:

Увеличить размер выборки: чем больше размер выборки, тем больше мощность теста.

Уменьшить уровень значимости: уменьшение уровня значимости позволяет увеличить мощность теста, но может увеличить вероятность ошибки первого рода.

Увеличить величину различия между группами: чем больше различие между группами, тем больше мощность теста.

Использовать более чувствительные тесты: некоторые статистические тесты более чувствительны, чем другие, поэтому использование более подходящего теста может увеличить мощность теста.

Знание мощности теста является важным при проектировании исследований, так как оно позволяет оценить, какой размер выборки необходим для достижения заданного уровня мощности. Кроме того, понимание мощности теста помогает интерпретировать результаты статистических тестов и принимать решения на основе этих результатов.

#### Минимально детектируемый эффект (Minimum Detectable Effect, MDE) - 
 
это наименьшее различие между двумя группами, которое можно обнаружить с заданным уровнем мощности и уровнем значимости.

Определение MDE является важным при проектировании исследований, так как оно позволяет определить минимальный размер эффекта, который следует ожидать, чтобы получить статистически значимые результаты при заданных параметрах статистического теста (уровне мощности и уровне значимости).

Например, если исследователь планирует использовать T-тест для сравнения двух групп, то он может определить MDE, чтобы определить, какой размер различия между группами следует ожидать, чтобы получить статистически значимые результаты с установленными уровнем мощности и уровнем значимости.

Определение MDE зависит от нескольких факторов, включая размер выборки, дисперсию данных и тип статистического теста, используемого для анализа данных. Чем больше размер выборки, тем меньше MDE, то есть можно обнаружить более маленькие различия между группами. Также MDE может быть уменьшен, используя более чувствительные статистические тесты.

Определение MDE - это важный этап в проектировании исследований, так как позволяет оценить, насколько точно исследование сможет обнаружить различия между группами и определить, достаточно ли большой размер выборки для достижения заданного уровня мощности и уровня значимости.

### Z - тест для пропорций

Z-тест для пропорций — это статистический тест, который используется для сравнения пропорций (или долей) двух независимых выборок. Он основан на нормальном распределении и используется, когда объемы выборок достаточно велики, чтобы применить центральную предельную теорему.

Для проведения z-теста для пропорций выполняются следующие шаги:

1. Рассчитать пропорцию для каждой группы. Пропорция — это отношение количества "успехов" (например, покупок) к общему количеству испытуемых (например, посетителей) в каждой группе.

2. Рассчитать объединенную пропорцию, которая представляет собой общую долю "успехов" в обеих группах. Это делается путем сложения "успехов" в обеих группах и деления на общее количество испытуемых в обеих группах.

3. Рассчитать разницу между пропорциями двух групп.

4. Рассчитать стандартное отклонение разности пропорций с использованием объединенной пропорции и размеров обеих выборок. Здесь используется предположение о том, что обе выборки взяты из одной и той же генеральной совокупности, и что пропорции в ней одинаковы.

5. Рассчитать z-статистику, которая представляет собой отношение разницы пропорций к стандартному отклонению разности пропорций. Z-статистика показывает, насколько далеко разница между пропорциями отклоняется от нуля в единицах стандартного отклонения.

6. Рассчитать p-значение на основе z-статистики. P-значение представляет вероятность получить такую или еще более выраженную разницу между пропорциями при условии, что в генеральной совокупности разницы между пропорциями нет (нулевая гипотеза). Если p-значение меньше выбранного уровня значимости (обычно 0.05), то можно сделать вывод о статистически значимых различиях между группами и отклонить нулевую гипотезу.

```python
# Данные о смертности после аварии
# среди тех кто пристегивался и нет
                 Выжили       Не выжили
Не пристегнул     1781        135
Пристегнул        1443        47
```

```python
from math import sqrt
from scipy.stats import norm
import statsmodels.api as sm

s1 = 135             # Число успехов выборка А
n1 = 1781 + 135      # Число наблюдений выборка А
s2 = 47              # Число успехов выборка B
n2 = 1443 + 47       # Число наблюдений выборка B

p1 = s1 / n1         # Доля успехов выборка А
p2 = s2 / n2         # Доля успехов выборка B

p = (s1 + s2) / (n1 + n2)  # Доля успехов А + В
z = (p2 - p1) / sqrt(p * (1 - p) * (1/n1 + 1/n2)) # Z метка

p_value = norm.cdf(z) * 2


z, p_value
```

```python
z, p_value = sm.stats.proportions_ztest([s1, s2], [n1, n2])

z, p_value
```

### Корреляционный анализ

Корреляционный анализ - это метод статистического анализа, который используется для изучения отношения между двумя переменными. Он позволяет определить, насколько сильно связаны две переменные и какой тип связи между ними.

Корреляционный анализ используется для изучения как линейных, так и нелинейных связей между двумя переменными. Он может быть использован для определения направления связи (положительной или отрицательной) и силы связи между двумя переменными. Коэффициент корреляции измеряет степень линейной связи между двумя переменными и может принимать значения от -1 до 1. Значение 1 указывает на положительную линейную связь, значение -1 указывает на отрицательную линейную связь, а значение 0 указывает на отсутствие линейной связи.

Корреляционный анализ может быть использован для многих целей, таких как:

Исследование связи между двумя переменными
Предсказание значения одной переменной на основе значения другой переменной
Определение факторов, влияющих на переменную
Проверка гипотезы о наличии связи между переменными
В Python для выполнения корреляционного анализа можно использовать библиотеки, такие как NumPy, pandas, и scipy. Например, чтобы рассчитать коэффициент корреляции между двумя переменными, можно использовать функцию corrcoef() из библиотеки NumPy:

```python
# создание данных с положительным коэффициентом корреляции
x = np.linspace(0, 10, 20)
y = 2 * x + np.random.randn(20)

# вычисление коэффициента корреляции
corr_coef = np.corrcoef(x, y)[0, 1]

# вывод результата
print(f"Correlation coefficient: {corr_coef:.2f}")

# построение графика
plt.scatter(x, y)
plt.title("Positive correlation")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# создание данных с отрицательным коэффициентом корреляции
x = np.linspace(0, 10, 20)
y = -2 * x + np.random.randn(20)

# вычисление коэффициента корреляции
corr_coef = np.corrcoef(x, y)[0, 1]

# вывод результата
print(f"Correlation coefficient: {corr_coef:.2f}")

# построение графика
plt.scatter(x, y)
plt.title("Negative correlation")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# создание данных с коэффициентом корреляции, равным нулю
x = np.linspace(0, 10, 20)
y = np.random.uniform(0, 10, 20)

# вычисление коэффициента корреляции
corr_coef = np.corrcoef(x, y)[0, 1]

# вывод результата
print(f"Correlation coefficient: {corr_coef:.2f}")

# построение графика
plt.scatter(x, y)
plt.title("Zero correlation")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## ANOVA

ANOVA (англ. Analysis of Variance) - это статистический метод, который используется для определения того, есть ли статистически значимые различия между двумя или более группами. Он позволяет определить, насколько различаются средние значения в разных группах и насколько вероятно это различие. ANOVA может использоваться для сравнения средних значений в трех или более группах.

Допустим, у нас есть три группы студентов, которые проходят разные методики обучения (метод A, метод B и метод C). Мы хотим сравнить эффективность этих методик обучения, используя их средний балл за финальный экзамен. Мы можем использовать однофакторный дисперсионный анализ (ANOVA) для определения, есть ли статистически значимые различия между средними баллами студентов в разных группах. Вот данные по каждой группе:

- Метод A: 75, 85, 93, 82, 90
- Метод B: 70, 65, 78, 75, 72
- Метод C: 80, 85, 90, 87, 88

```python
import numpy as np
from scipy.stats import f_oneway

# Данные оценок студентов для каждого метода обучения
method_a = np.array([75, 85, 93, 82, 90])
method_b = np.array([70, 65, 78, 75, 72])
method_c = np.array([80, 85, 90, 87, 88])

# Применяем однофакторный дисперсионный анализ (ANOVA)
f_statistic, p_value = f_oneway(method_a, method_b, method_c)

# Выводим результаты
print("F-статистика:", f_statistic)
print("p-значение:", p_value)

# Определение статистической значимости (например, при уровне значимости 0.05)
alpha = 0.05
if p_value < alpha:
    print("Есть статистически значимые различия между группами.")
else:
    print("Нет статистически значимых различий между группами.")
```

```python
# Объединяем данные для создания DataFrame
data = np.concatenate((method_a, method_b, method_c))
labels = ["Method A"] * len(method_a) + ["Method B"] * len(method_b) + ["Method C"] * len(method_c)

df = pd.DataFrame({"Method": labels, "Scores": data})

# Создаем график boxplot с использованием seaborn
plt.figure(figsize=(8, 6))
sns.boxplot(x="Method", y="Scores", data=df)
plt.title("Scores by Teaching Method")
plt.xlabel("Teaching Method")
plt.ylabel("Scores")
```

```python
# Создаем график pointplot с использованием seaborn
plt.figure(figsize=(8, 6))
sns.pointplot(x="Method", y="Scores", data=df, capsize=.1)
plt.title("Scores by Teaching Method with Confidence Intervals")
plt.xlabel("Teaching Method")
plt.ylabel("Scores")

plt.show()
```

A/B-тестирование — это неотъемлемая часть процесса работы над продуктом. Это эксперимент, который позволяет сравнить две версии чего-либо, чтобы проверить гипотезы и определить, какая версия лучше. Должны ли кнопки быть черными или белыми, какая навигация лучше, какой порядок прохождения регистрации меньше всего отпугивает пользователей? Продуктовый дизайнер из Сан-Франциско Лиза Шу рассказывает о простой последовательности шагов, которые помогут провести базовое тестирование.

### Кому нужно A/B-тестирование
— Продакт-менеджеры могут тестировать изменения ценовых моделей, направленные на повышение доходов, или оптимизацию части воронки продаж для увеличения конверсии.

— Маркетологи могут тестировать изображения, призывы к действию (call-to-action) или практически любые другие элементы маркетинговой кампании или рекламы с точки зрения улучшения метрик.

— Продуктовые дизайнеры могут тестировать дизайнерские решения (например, цвет кнопки оформления заказа) или использовать результаты тестирования для того, чтобы перед внедрением определить, будет ли удобно пользоваться новой функцией.

Вот шесть шагов, которые нужно пройти, чтобы провести тестирование. В некоторые из пунктов включены примеры тестирования страницы регистрации выдуманного стартапа.

### 1. Определите цели 
Определите основные бизнес-задачи вашей компании и убедитесь, что цели A/B-тестирования с ними совпадают.

Пример: Допустим, вы менеджер продукта в «компании X» на стадии стартапа. Руководству нужно добиться роста количества пользователей. В частности, компания стремится к росту количества активных пользователей (метрика DAU), определяемых как среднее количество зарегистрированных пользователей сайта в день за последние 30 дней. Вы предполагаете, что этого можно добиться либо путем улучшения показателей удержания (процент пользователей, возвращающихся для повторного использования продукта), либо путем увеличения числа новых регистрирующихся пользователей.

В процессе исследования воронки вы замечаете, что 60% пользователей уходят до завершения регистрации. Это означает, что можно повысить количество регистраций, изменив страницу регистрации, что, в свою очередь, должно помочь увеличить количество активных пользователей.

### 2. Определите метрику 
Затем вам нужно определить метрику, на которую вы будете смотреть, чтобы понять, является ли новая версия сайта более успешной, чем изначальная. Обычно в качестве такой метрики берут коэффициент конверсии, но можно выбрать и промежуточную метрику вроде показателя кликабельности (CTR).

Пример: В нашем примере в качестве метрики вы выбираете долю зарегистрированных пользователей (registration rate), определяемую как количество новых пользователей, которые регистрируются, поделенное на общее количество новых посетителей сайта.

### 3. Разработайте гипотезу 
Затем нужно разработать гипотезу о том, что именно поменяется, и, соответственно, что вы хотите проверить. Нужно понять, каких результатов вы ожидаете и какие у них могут быть обоснования.

Пример: Допустим, на текущей странице регистрации есть баннер и форма регистрации. Есть несколько пунктов, которые вы можете протестировать: поля формы, позиционирование, размер текста, но баннер на главной странице визуально наиболее заметен, поэтому сначала надо узнать, увеличится ли доля регистраций, если изменить изображение на нём.

Общая гипотеза заключается в следующем: «Если изменить главную страницу регистрации, то больше новых пользователей будут регистрироваться внутри продукта, потому что новое изображение лучше передает его ценности».

Нужно определить две гипотезы, которые помогут понять, является ли наблюдаемая разница между версией A (изначальной) и версией B (новой, которую вы хотите проверить) случайностью или результатом изменений, которые вы произвели.

— Нулевая гипотеза предполагает, что результаты, А и В на самом деле не отличаются и что наблюдаемые различия случайны. Мы надеемся опровергнуть эту гипотезу.

— Альтернативная гипотеза — это гипотеза о том, что B отличается от A, и вы хотите сделать вывод об её истинности.

Решите, будет ли это односторонний или двусторонний тест. Односторонний тест позволяет обнаружить изменение в одном направлении, в то время как двусторонний тест позволяет обнаружить изменение по двум направлениям (как положительное, так и отрицательное).  

### 4. Подготовьте эксперимент 
Для того, чтобы тест выдавал корректные результаты сделайте следующее:

— Создайте новую версию (B), отражающую изменения, которые вы хотите протестировать.

— Определите контрольную и экспериментальную группы. Каких пользователей вы хотите протестировать: всех пользователей на всех платформах или только пользователей из одной страны? Определите группу испытуемых, отобрав их по типам пользователей, платформе, географическим показателям и т. п. Затем определите, какой процент исследуемой группы составляет контрольная группа (группа, видящая версию A), а какой процент — экспериментальная группа (группа, видящая версию B). Обычно эти группы одинакового размера.

— Убедитесь, что пользователи будут видеть версии A и B в случайном порядке. Это значит, у каждого пользователя будет равный шанс получить ту или иную версию.

— Определите уровень статистической значимости (α). Это уровень риска, который вы принимаете при ошибках первого рода (отклонение нулевой гипотезы, если она верна), обычно α = 0.05. Это означает, что в 5% случаев вы будете обнаруживать разницу между A и B, которая на самом деле обусловлена случайностью. Чем ниже выбранный вами уровень значимости, тем ниже риск того, что вы обнаружите разницу, вызванную случайностью.

— Определите минимальный размер выборки. Ссылки на калькуляторы есть выше в ноутбуке, они рассчитывают размер выборки, необходимый для каждой версии. На размер выборки влияют разные параметры и ваши предпочтения. Наличие достаточно большого размера выборки важно для обеспечения статистически значимых результатов.

— Определите временные рамки. Возьмите общий размер выборки, необходимый вам для тестирования каждой версии, и разделите его на ваш ежедневный трафик, так вы получите количество дней, необходимое для проведения теста. Как правило, это одна или две недели.

Пример: На существующем сайте в разделе регистрации мы изменим главную страницу — это и будет нашей версией B. Мы решаем, что в эксперименте будут участвовать только новые пользователи, заходящие на страницу регистрации. Мы также обеспечиваем случайную выборку, то есть каждый пользователь будет иметь равные шансы получить A или B, распределенные случайным образом.

Важно определить временные рамки. Допустим, ежедневно на нашу страницу регистрации в среднем приходит трафик от 10 000 новых пользователей, это означает, что только 5000 пользователей могут увидеть каждую версию. Тогда минимальный размер выборки составляет около 100 000 просмотров каждой версии. 100 000/ 5000 = 20 дней — столько должен продлиться эксперимент.

### 5. Проведите эксперимент 
Помните о важных шагах, которые необходимо выполнить:

— Обсудите параметры эксперимента с исполнителями.

— Выполните запрос на тестовой закрытой площадке, если она у вас есть. Это поможет проверить данные. Если ее нет, проверьте данные, полученные в первый день эксперимента.

— В самом начале проведения тестирования проверьте, действительно ли оно работает.

— И, наконец, не смотрите на результаты! Преждевременный просмотр результатов может испортить статистическую значимость.

### 6. Анализируйте результаты. Наконец-то самое интересное 
Вам нужно получить данные и рассчитать значения выбранной ранее метрики успеха для обеих версий (A и B) и разницу между этими значениями. Если не было никакой разницы в целом, вы также можете сегментировать выборку по платформам, типам источников, географическим параметрам и т. п., если это применимо. Вы можете обнаружить, что версия B работает лучше или хуже для определенных сегментов.

Проверьте статистическую значимость. Статистическая теория, лежащая в основе этого подхода, объясняется здесь, но основная идея в том, чтобы выяснить, была ли разница в результатах между A и B связана с изменениями или это результат случайности или естественных изменений. Это определяется путем сравнения тестовых статистических данных (и полученного p-значения) с вашим уровнем значимости.

Если p-значение меньше уровня значимости, то можно отвергнуть нулевую гипотезу, имея доказательства для альтернативы.

Если p-значение больше или равно уровню значимости, мы не можем отвергнуть нулевую гипотезу о том, что A и B не отличаются друг от друга.

A/B-тестирование может дать следующие результаты:

— Контрольная версия, А выигрывает или между версиями нет разницы. Если исключить причины, которые могут привести к недействительному тестированию, то проигрыш новой версии может быть вызван, например, плохим сообщением и брендингом конкурентного предложения или плохим клиентским опытом.

В этом сценарии вы можете углубиться в данные или провести исследование пользователей, чтобы понять, почему новая версия не работает так, как ожидалось. Это, в свою очередь, поможет собрать информацию для следующих тестов.

— Версия B выигрывает. A/B-тест подтвердил вашу гипотезу о лучшей производительности версии B по сравнению с версией A. Отлично! Опубликовав результаты, вы можете провести эксперимент на всей аудитории и получить новые результаты.

### Заключение
Независимо от того, был ли ваш тест успешным или нет, относитесь к каждому эксперименту как к возможности для обучения. Используйте то, чему вы научились, для выработки вашей следующей гипотезы. Вы можете, например, использовать предыдущий тест или сконцентрироваться на другой области, требующей оптимизации. Возможности бесконечны.

### Типичный пример A/B теста на python

Интернет-магазин внедрил новый дизайн страницы продукта. Вы хотите проверить, увеличивает ли новый дизайн конверсию покупок по сравнению со старым дизайном. Вы провели A/B-тест и собрали данные о количестве пользователей и количестве покупок для каждой версии страницы продукта.

Для простоты предположим, что у нас есть следующие данные:

Группа A (старый дизайн):

Посетителей: 1200
Покупок: 110
Группа B (новый дизайн):

Посетителей: 1000
Покупок: 125
